Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

Loading Dataset

```python
df = pd.read_csv("spg.csv")

df.head()
```

DATA CLEANING

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

Removing Duplicate Rows

```python
df = df.drop_duplicates()
df = df[df['generated_power_kw'] >= 0]
```

Outlier Removal (IQR Method)

```python
Q1 = df['generated_power_kw'].quantile(0.25)
Q3 = df['generated_power_kw'].quantile(0.75)
IQR = Q3 - Q1

df = df[
    (df['generated_power_kw'] >= Q1 - 1.5 * IQR) &
    (df['generated_power_kw'] <= Q3 + 1.5 * IQR)
]
```

Feature Selection

```python
features = [
    'generated_power_kw',
    'shortwave_radiation_backwards_sfc',
    'temperature_2_m_above_gnd',
    'total_cloud_cover_sfc',
    'zenith',
    'wind_speed_10_m_above_gnd'
]

df_clean = df[features]
```

Cleaned Data checking

```python
df_clean.describe()
```

Average Power by Cloud Cover

```python
df['cloud_category'] = pd.cut(
    df['total_cloud_cover_sfc'],
    bins=[0, 30, 60, 100],
    labels=['Low', 'Medium', 'High']
)

avg_power = df.groupby('cloud_category')['generated_power_kw'].mean()

avg_power.plot(kind='bar')

plt.xlabel("Cloud Cover Level")
plt.ylabel("Average Power Generated (kW)")
plt.title("Average Solar Power by Cloud Cover")
plt.show()
```

Solar Radiation vs Power Trend

```python
df_sorted = df.sort_values(
    by='shortwave_radiation_backwards_sfc'
)

plt.plot(
    df_sorted['shortwave_radiation_backwards_sfc'],
    df_sorted['generated_power_kw']
)

plt.xlabel("Solar Radiation")
plt.ylabel("Generated Power (kW)")
plt.title("Trend of Power Generation with Solar Radiation")
plt.show()
```

Temperature vs Power Distribution

```python
sns.boxplot(
    x=df['temperature_2_m_above_gnd'],
    y=df['generated_power_kw']
)

plt.xlabel("Temperature (°C)")
plt.ylabel("Generated Power (kW)")
plt.title("Power Generation Distribution Across Temperature")
plt.show()
```

Radiation vs Power (Minimal)

```python
sns.scatterplot(
    x='shortwave_radiation_backwards_sfc',
    y='generated_power_kw',
    data=df,
    alpha=0.6
)

plt.xlabel("Solar Radiation")
plt.ylabel("Generated Power (kW)")
plt.title("Solar Radiation vs Power Generation")
plt.show()
```
